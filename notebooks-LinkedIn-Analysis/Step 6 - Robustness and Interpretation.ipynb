## Report Visual Aids

These robustness visuals compare the saved variant outputs directly, so they can be run without rebuilding the alternative corpora or rerunning WEAT.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path

group_order = ["entry", "mid", "senior", "leadership"]
robustness_dir = Path("../results/robustness")

lexicon_variants = {
    "Deduped + backfilled": pd.read_csv(robustness_dir / "linkedin_digital_tech_lexicon_bias_labeled_plus_backfilled.csv"),
    "Labeled-only": pd.read_csv(robustness_dir / "linkedin_digital_tech_lexicon_bias_labeled_only.csv"),
    "Non-deduped": pd.read_csv(robustness_dir / "linkedin_digital_tech_lexicon_bias_nondedup.csv"),
}
broad_summary = pd.read_csv(robustness_dir / "linkedin_digital_tech_broader_tech_company_summary.csv")
weat_main = pd.read_csv(Path("../results/weat/weat_summary.csv"))
weat_labeled_only = pd.read_csv(Path("../results/weat_labeled_only/weat_summary.csv"))

lexicon_long = []
for variant_name, df in lexicon_variants.items():
    variant_df = df.copy()
    variant_df["variant"] = variant_name
    lexicon_long.append(variant_df)
lexicon_long = pd.concat(lexicon_long, ignore_index=True)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for variant_name, subset in lexicon_long.groupby("variant"):
    ordered = subset.set_index("group").reindex(group_order)
    axes[0, 0].plot(group_order, ordered["bias_score"], marker="o", linewidth=2, label=variant_name)
axes[0, 0].axhline(0, color="black", linewidth=1)
axes[0, 0].set_title("Lexicon bias score across robustness variants")
axes[0, 0].set_ylabel("Masculine - feminine per 1,000 tokens")
axes[0, 0].legend(title="", fontsize=9)

width = 0.25
x = np.arange(len(group_order))
for offset, (variant_name, df) in zip([-width, 0, width], lexicon_variants.items()):
    ordered = df.set_index("group").reindex(group_order)
    axes[0, 1].bar(x + offset, ordered["token_total"], width=width, label=variant_name)
axes[0, 1].set_xticks(x)
axes[0, 1].set_xticklabels(group_order)
axes[0, 1].set_title("Token volume behind each robustness variant")
axes[0, 1].set_ylabel("Tokens")
axes[0, 1].legend(title="", fontsize=9)

compare = pd.DataFrame({"group": group_order})
compare = compare.merge(weat_main[["group", "effect_size_mean"]].rename(columns={"effect_size_mean": "Main WEAT"}), on="group", how="left")
compare = compare.merge(
    weat_labeled_only[["group", "effect_size_mean"]].rename(columns={"effect_size_mean": "Labeled-only WEAT"}),
    on="group",
    how="left",
)
axes[1, 0].bar(x - width / 2, compare["Main WEAT"], width=width, label="Main WEAT", color="#4c72b0")
axes[1, 0].bar(x + width / 2, compare["Labeled-only WEAT"].fillna(0), width=width, label="Labeled-only WEAT", color="#55a868")
for idx, value in enumerate(compare["Labeled-only WEAT"]):
    if pd.isna(value):
        axes[1, 0].text(x[idx] + width / 2, 0.03, "NA", ha="center", va="bottom", color="#c44e52", fontsize=9)
axes[1, 0].axhline(0, color="black", linewidth=1)
axes[1, 0].set_xticks(x)
axes[1, 0].set_xticklabels(group_order)
axes[1, 0].set_title("WEAT robustness: main vs labeled-only")
axes[1, 0].set_ylabel("Effect size")
axes[1, 0].legend(title="", fontsize=9)

broad_plot = broad_summary.set_index("metric").loc[["digital_tech_rows", "broad_tech_company_rows"], "value"]
broad_plot.index = ["Digital tech rows", "Broader tech-company rows"]
axes[1, 1].bar(broad_plot.index, broad_plot.values, color=["#55a868", "#8172b3"])
axes[1, 1].set_title("Scope robustness benchmark")
axes[1, 1].set_ylabel("Rows")
axes[1, 1].tick_params(axis="x", rotation=15)

plt.tight_layout()
plot_name = "Step 6 - Report Visual Aids"
plt.savefig(f"{plot_name}.png", dpi=300, bbox_inches="tight")
plt.show()
```
